```python

# Classes & Objects

class Dog:
    species : str = "Canis familaris"
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def bark(self):
        return f"{self.name} says Woof!"
    
my_dog = Dog("Max", 3)
print(my_dog.name);
print(my_dog.age);
print(my_dog.bark());

# Inheritance

class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
       raise NotImplementedError("Subclass must implement this method")

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"  

class Duck(Animal):
    def speak(self):
        return f"{self.name} says Quack!"    
    
my_cat = Cat("Whiskers");
print(my_cat.speak());

my_duck = Duck("Daisy");
print(my_duck.speak());

# Encapsulation 

class BankAccount:
    def __init__(self, balance):
       self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance
        
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance());

# Polymorphism

class Bird:
    def fly(self):
        return "I can fly!"

class Penguin(Bird):
    def fly(self):
        return "I can't fly, but i can swim!"   

bird = Bird()
penguin = Penguin()

print(bird.fly());
print(penguin.fly());
```

```python
class savingsAccount(bankAccount):
    def __init__(self, savings, duration, interest):
        self.savings = savings
        self.duration = duration
        self.interest = interest
        
    def apply_interest(self, balance):
        if self.savings >= balance and self.duration >= 6:
            print(f"interest added : ${self.interest}");
        else:
            print(f"Save to get up to ${self.interest} Interest for 6months and above");

interest = savingsAccount(2000, 4, 500);
interest.apply_interest(balance= 2000);
```
